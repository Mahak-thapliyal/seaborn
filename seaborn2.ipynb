```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
var=sns.load_dataset("penguins")
var
```

```python
sns.displot(var["flipper_length_mm"],log_scale=True)
plt.show()
```

```python
sns.displot(var["flipper_length_mm"],bins=[170,180,190,200,210,220,230,240],kde=True,rug=True,color="coral") #kernel density estimator
plt.show()
```

```python
#scatter plot
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
var=sns.load_dataset("penguins")
m={"Male":"*","Female":"o"}
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=var,hue="sex",style="sex",size="sex",sizes=(30,60),palette="plasma",markers=m)
plt.show()
```
