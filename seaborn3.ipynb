```python
#violin plot
#A violin plot plays a similar role as a box and whisker plot.It shows the distribution of quantitative data across several levels of one(or more)categorical variables such that those distributions can be compared.
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
var=sns.load_dataset("tips")
var
```

```python
sns.violinplot(x="day",y="total_bill",data=var,linewidth=3,saturation=0.1)
```

```python
sns.violinplot(x="day",y="total_bill",data=var,hue="time",palette="vlag")
```

```python
sns.violinplot(x="time",y="total_bill",data=var,order=["Dinner","Lunch"],inner="quart")
#inner parameter can have box,quartile,point,stick,none
```

```python
sns.violinplot(x="day",y="total_bill",data=var,hue="sex",split=True,scale="count")
#scale parameter can have count,width,area.
```

```python
sns.violinplot(y="day",x="total_bill",data=var,hue="sex",split=True,scale="count")
```

```python
sns.violinplot(x=var["total_bill"])
```

```python
#Pair plot
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
var=sns.load_dataset("tips")
var
```

```python
sns.pairplot(var)
plt.show()
```

```python
sns.pairplot(var,hue="sex")
plt.show()
```

```python
sns.pairplot(var,vars=["total_bill","tip"],hue="sex",hue_order=['Female','Male'])
plt.show()
```

```python
sns.pairplot(var,y_vars=["total_bill","tip"])
plt.show()
```

```python
sns.pairplot(var,kind="reg") #kind parameter values(scatter,kde,hist)
plt.show()
```

```python
sns.pairplot(var,kind="kde",diag_kind="hist") 
plt.show()
```

```python
sns.pairplot(var,vars=["total_bill","tip"],hue="sex",hue_order=['Female','Male'],markers=["*","<"])
plt.show()
```

```python
#strip plot
#A strip plot is very simple to understand,it is basically a scatter plot that differentiates a different categories.Strip plots are considered a good alternative to a box plot or a violin plot.
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
var=sns.load_dataset("tips")
var
```

```python

sns.stripplot(x="day",y="total_bill",data=var,hue="sex",palette="rocket_r",edgecolor="b",alpha=0.4)
```

```python
sns.stripplot(x="day",y="total_bill",data=var,hue="sex",palette="rocket_r",edgecolor="b",jitter=5,size=3,marker="<")
```

```python
sns.stripplot(x=var["total_bill"])
```

```python
#box plot
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
var=sns.load_dataset("tips")
var
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
var=sns.load_dataset("tips")
sns.boxplot(y=var["total_bill"])
```

```python
sns.set_style("darkgrid")
sns.barplot(x="day",y="total_bill",data=var)

plt.show()
```

```python
sns.set_style("white")
sns.set_context("talk",font_scale=1) #(poster,paper,talk)
plt.figure(figsize=(4,6))
sns.barplot(x="day",y="total_bill",data=var)
sns.despine()

plt.show()
```

```python
#multiple plots(Facet-grid)
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
var=sns.load_dataset("tips")
fg=sns.FacetGrid(var,col="sex").add_legend()
fg.map(plt.scatter,"total_bill","tip")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
var=sns.load_dataset("tips")
fg=sns.FacetGrid(var,col="day").add_legend()
fg.map(plt.bar,"total_bill","tip")
```
