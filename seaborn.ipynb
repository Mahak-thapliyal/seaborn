```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
var=sns.load_dataset("penguins") # data from git hub 
var
```

```python
sns.barplot(x=var.island,y=var.bill_length_mm)
```

```python
sns.barplot(x=var.island,y=var.bill_length_mm)
plt.show()
```

```python
sns.barplot(x="island",y="bill_length_mm",data=var)
plt.show()
```

```python
sns.barplot(x="island",y="bill_length_mm",data=var,hue="sex")
plt.show()
```

```python
order_1=["Dream","Torgersen","Biscoe"]
sns.barplot(x="island",y="bill_length_mm",data=var,hue="sex",order=order_1,hue_order=["Female","Male"],n_boot=9,palette="plasma",saturation=0.3,errcolor="b",errwidth=4,capsize=1)
plt.show()
```

```python
#horizontal bar graph should have both numeric parameter x and y should be numerical
order_1=["Dream","Torgersen","Biscoe"]
sns.barplot(x="bill_depth_mm",y="bill_length_mm",data=var,orient="h")
plt.show()
```

```python
order_1=["Dream","Torgersen","Biscoe"]
sns.barplot(x="island",y="bill_length_mm",data=var,color="g")
plt.show()
```

```python
sns.set(style="darkgrid")
order_1=["Dream","Torgersen","Biscoe"]
sns.barplot(x="island",y="bill_length_mm",data=var,hue="sex",order=order_1,hue_order=["Female","Male"],n_boot=9,palette="plasma",saturation=0.3,errcolor="b",dodge=False,alpha=0.9)
plt.show()
```
